## Bias and Variance

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings

warnings.filterwarnings("ignore")
```

```python
# Dataset
X = np.array([1, 2, 3]).reshape(-1, 1)
y = np.array([1, 4, 9])  
#True relationship  y = x^2
```

## Train multiple models
- Linear model
- High-degree polynomial 

## Model-1

```python
model_1 = LinearRegression()
model_1.fit(X, y)
print("Slope:", model_1.coef_)
print("Intercept:", model_1.intercept_)
```

```python
print("Model_1 Predictions:", model_1.predict(X))
```

## Model-2

```python
y=ax^5+bx^4+cx^3+dx^2+ex+f
```

```python
x = np.array([1, 2, 3])
y = np.array([1, 4, 9])

# coefficients 
coeff = np.polyfit(x, y, 5)

print(coeff)
```

```python
poly = np.poly1d(coeff)
y_pred_train = poly(x)
print("Predictions on training data:", y_pred_train)
```

```python
x_test = 4
y_true = 16  
y_pred_test = poly(x_test)
print("Prediction at x=4:", y_pred_test)
print("True value:", y_true)
```

# PROBLEM 1

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings("ignore")
```

```python
X = np.array([1, 2, 3,4,5]).reshape(-1, 1)
y = np.array([52,55,61,70,82])
```

# Model1

```python
model_1 = LinearRegression()
model_1.fit(X,y)
print("Slope:", model_1.coef_)
print("Intercept:", model_1.intercept_)
```

```python
y_pred_linear = model_1.predict(X)


int1 = model_1.intercept_
int2 = model_1.coef_[0]

print("Model 1: Simple Linear Regression")
print(f"Equation: y = {int1:.2f} + {int2:.2f}x")

mse_linear = mean_squared_error(y, y_pred_linear)
print(f"Training MSE (Linear Model): {mse_linear:.4f}")

X_test = np.array([[6]])
print(f"Prediction at x=6: {model_1.predict(X_test)[0]:.2f}")
```

# Model 2

```python
model_2 = PolynomialFeatures(degree=4)
x_poly = model_2.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(x_poly, y)
```

```python
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(X)

model_2 = LinearRegression()
model_2.fit(x_poly, y)

y_pred_poly = model_2.predict(x_poly)

print("\nModel B: Polynomial Regression (Degree 4)")
print("Training MSE:", mean_squared_error(y, y_pred_poly))

x_test = poly.transform([[6]])
print("Prediction at x = 6:", model_2.predict(x_test)[0])
```

```python
print("Bias–Variance Reasoning")
print("Model A (Linear Regression):")
print("- Higher Bias")
print("- Reason: Too simple, underfits the non-linear relationship")

print("\nModel B (Polynomial Regression - Degree 4):")
print("- Higher Variance")
print("- Reason: Too complex, overfits the training data")
```
